```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import matplotlib.animation as animation
from IPython.display import HTML
```

```python
df = pd.read_excel('co2_emission_ext.xlsx')
dff = df.iloc[10:55]
dff
```

```python
colors={
    'Philippines [PHL]':'#ffe4b0',
    'Malaysia [MYS]':'#d4af37',
    'Indonesia [IDN]':'#edced8',
    'Thailand [THA]':'#ced8ed',
    'Singapore [SGP]':'#aa6f91',
    'Vietnam [VNM]':'#7dfe3f',
    'Cambodia [KHM]':'#fd837b'
}
```

```python


def draw_barchart(current_year):
    df2 = dff.loc[df['Time'] == current_year].set_index('Series Name').T
    df3 = df2.iloc[3:]['CO2 emissions (kt)']
    df4 = df3.sort_values(ascending=True)
    country = df4.index
    ax.clear()
    ax.barh(country, df4, color=[colors[i] for i in country])

    dv = df4.max()/100

    for i, (value, name) in enumerate(zip(df4, country)):
        ax.text(value-dv, i, name,size=20, weight=700, ha='right', va='center')
        ax.text(value+dv, i, f'{value:,.0f}', size=17, ha='left',  va='center')

    ax.text(.9, .2, current_year, transform=ax.transAxes, color='#777777', size=72, ha='right', weight=800)
    ax.text(.05, 1.25, 'CO2 Emissions of Southeast Asian Countries (1970-2014)',
            transform=ax.transAxes, size=36, weight=800, ha='left', va='top', name='calibri')
    ax.text(0, 1.1, 'CO2 emission in kt', transform=ax.transAxes, size=17, color='#777777')
    ax.xaxis.set_ticks_position('top')
    ax.tick_params(axis='x', colors='#777777', labelsize=17)
    ax.set_yticks([])
    ax.margins(0, 0.01)
    ax.grid(which='major', axis='x', linestyle='-')
    ax.set_axisbelow(True)
    plt.box(False)
```

```python
# or use animator.to_html5_video() or animator.save() 
```

```python
fig, ax = plt.subplots(figsize=(20, 8))

animator = animation.FuncAnimation(fig, draw_barchart, frames=range(1970, 2015), repeat=False)
animator.save('co2_bar_race.gif', writer='imagemagick', fps=30)
```
